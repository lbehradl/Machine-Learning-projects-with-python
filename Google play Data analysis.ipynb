## Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

## Load and Prepare Data

## Data Preprocessing

```python
df2 = df.dropna()
df2.info()
```

```python
df2 = df2.astype(str)

df2['Rating'] = df2['Rating'].astype('float')

df2['Reviews'] = df2['Reviews'].astype('int')

df2['Size'] = df2['Size'].replace('Varies with device', np.nan)
df2['Size'] = df2['Size'].str.replace('M', 'e6').str.replace('k', 'e3')
df2['Size'] = df2['Size'].astype(float)

df2['Installs'] = df2['Installs'].str.replace(',', '').str.replace('+','')
df2['Installs'] = df2['Installs'].astype('int')

df2['Price'] = df2['Price'].str.replace('$', '')
df2['Price'] = df2['Price'].astype(float)

df2['Content Rating'] = df2['Content Rating'].replace('Unrated', np.nan)
df2.info()
```

```python
df2 = df2.dropna()
df2.describe()
```

## Strorytelling - Visualization

```python
#نمودار رایگان یا پولی بودن نرم افزارها
    
sns.displot(df2, y='Type', bins=20, shrink = 0.5)
plt.title('Distribution of Type')
plt.show()
```

```python
#نمودار ریتینگ با تفکیک رایگان یا پولی بودن

sns.displot(df2, y='Rating', bins=20, hue='Type')
plt.title('Distribution of Ratings by Content Type')
plt.show()
```

```python
# نمودار ریتینگ با تفکیک سن مناسب 

sns.displot(df2, x='Rating',kind='kde', hue='Content Rating')
plt.title('Distribution of Ratings by Content Rating')
plt.show()
```

```python
# نمودار تعداد نصب با تفکیک سن

sns.displot(df2, x='Installs', bins=15, log_scale=True, hue='Content Rating')
plt.title('Logarithmic Distribution of Installs by Type')
plt.show()
```

```python
# نمودار تعداد نصب با تفکیک کتگوری


sns.displot(df2, x='Installs',  bins=15, hue='Category', log_scale=True)
plt.title('Logarithmic Distribution of Installs by Category')
plt.show()
```

```python
# نمودار ریتینگ براساس تعداد نظرات با تفکیک تایپ و بزرگنمایی تعداد نصب

sns.relplot(df2, x='Rating', y='Reviews', hue='Type', size='Installs',sizes=(40, 400),  aspect=2, palette='viridis')
plt.title('Relationship between Rating and Reviews by Type')
plt.show()
```

```python
# نمودار تعداد نصب نسبت به تمره و تفگیگ تایپ

sns.relplot(df2, x=np.log1p(df2['Installs']), y='Rating', hue='Type')
plt.title('Logarithmic Relationship between Installs and Rating')
plt.xlabel('Log(Installs)')
plt.show()
```

```python
# نمودار نمره براساس کتگوری 

plt.figure(figsize=(12, 8))

sns.boxplot(df2, x='Rating', y='Category')
plt.title('Distribution of Ratings by Category')
plt.show()
```
